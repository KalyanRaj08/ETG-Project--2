# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('summer.csv')
hf = df.dropna()
```

### 1. In how many cities Summer Olympics is held so far?

```python
len(df['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
gold=[]
gf = df[df['Medal'] == 'Gold']
for sport in gf['Sport'].unique():
  gold.append([sport , len(gf[gf['Sport'] == sport])])
pd.DataFrame(gold,columns=['Sport','No of Gold']).sort_values(by='No of Gold',ascending=False).head().plot(x='Sport' , y='No of Gold' , kind='bar' , figsize=(5,10))
```

```python
for i in gold:
  print (i[0],':',i[1])
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data=[]
for sport in df['Sport'].unique():
  data.append([sport , len(df[df['Sport'] == sport])])
pd.DataFrame(data,columns=['Sport','Medals']).sort_values(by='Medals',ascending=False).head().plot(x='Sport' , y='Medals' , kind='bar' , figsize=(5,10))
```

```python
for i in data:
  print (i[0],':',i[1])
```

### 4. Which player has won most number of medals? (Top 5)

```python
plyr=[]
for player in df['Athlete'].unique():
  plyr.append([player , len(df[df['Athlete'] == player])])
pd.DataFrame(plyr , columns=['Player' , 'No of medals']).sort_values(by = 'No of medals' ,ascending = False).head().plot(x='Player' , y= 'No of medals', kind = 'bar', figsize = (5,10))
```

```python
for i in plyr:
  print (i[0],':',i[1])
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data = []
for i in df['Event'].unique():
  gf = df[df['Event'] == i]
  data.append([i , len(gf['Athlete'].unique())])
pd.DataFrame(data, columns=['Event', 'No of Players']).sort_values(by= 'No of Players', ascending= False).head().plot(x= 'Event', y= 'No of Players', kind= 'bar', figsize = (10,10))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
data=[]
gf = df[df['Gender'] == 'Women']
cf = gf[gf['Medal'] == 'Gold']
for i in cf['Sport'].unique():
  data.append([i, len(cf[cf['Sport'] == i])])
pd.DataFrame(data, columns=['Sport', 'No of Players']).sort_values(by= 'No of Players', ascending= False).head().plot(x= 'Sport', y= 'No of Players', kind= 'bar', figsize = (10,10))
```
